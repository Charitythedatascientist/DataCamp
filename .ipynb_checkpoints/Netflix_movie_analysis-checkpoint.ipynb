```python
#To get the names of the columns in the dataset
netflix_df.columns
```

```python
#To get tv shows released from 2018 in Brazil
tv_shows_brazil = netflix_df[(netflix_df["country"] == 'Brazil')& (netflix_df["release_year"] > 2018) & (netflix_df["type"] == "TV Show")]
```

```python
tv_shows_brazil.head()
```

```python
#Remove tv shows from the dataframe.
netflix_subset = netflix_df[netflix_df["type"] != "TV Show"]
```

```python
netflix_subset.head()
```

```python
#To see unique values in the type column
netflix_df["type"].nunique()
```

```python
#keep "title", "country", "genre", "release_year", "duration", and saving this into a new DataFrame called netflix_movies.
netflix_movies = netflix_df[["title", "country", "genre", "release_year", "duration"]]
```

```python
netflix_movies.head()
```

```python
#netflix_movies shorter than 60 minutes
short_movies = netflix_df[netflix_df["duration"] < 60]
```

```python
short_movies.head()
```

```python
#loop through the dataset, and add different genres as colors in a list.
colors = []
for lab, row in netflix_movies.iterrows():
    if row['genre'] == 'Children':
        colors.append("Red")
    elif row['genre'] == 'Documentaries':
        colors.append('Blue')
    elif row['genre'] == 'Stand-Up':
        colors.append('Yellow')
    else:
        colors.append('Black')
```
